```python
df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")
```

```python
df.head()
```

```python
conn = sqlite3.connect("superstore.db")
```

```python
df.to_sql(
    "superstore",
    conn,
    if_exists="replace",
    index=False
)
```

```python
def run_query(sql):
    return pd.read_sql_query(sql, conn)
```

```python
# Retrive all customer information
run_query("""
SELECT * FROM superstore LIMIT 5;
""")
```

```python
# Check total number of records in the dataset

run_query("""
SELECT COUNT(*) AS Total_Records FROM superstore;
""")
```

```python
# Find all unique product categories

run_query("""
SELECT DISTINCT "Category"
FROM superstore;
""")
```

```python
# Filter orders from West region using WHERE

run_query("""
SELECT * FROM superstore
WHERE "Region" = 'West';
""")
```

```python
# Orders with sales between 100 and 500

run_query("""
SELECT * FROM superstore
WHERE Sales BETWEEN 100 AND 500;
""")
```

```python
# Orders from East and West regions using IN operator

run_query("""
SELECT *
FROM superstore
WHERE "Region" IN ('East','West');
""")
```

```python
# Customers whose names start with A

run_query("""
SELECT DISTINCT "Customer Name"
FROM superstore
WHERE "Customer Name" LIKE 'A%';
""")
```

```python
# Top 10 highest sales

run_query("""
SELECT *
FROM superstore
ORDER BY Sales DESC
LIMIT 10;
""")
```

```python
# Total sales by category

run_query("""
SELECT "Category",SUM(sales) AS Total_Sales
FROM superstore
GROUP BY "Category";
""")
```

```python
# Average profit by region

run_query("""
SELECT
    "Region",Avg(profit) AS Avg_Profit
FROM superstore
GROUP BY "Region";
""")
```

```python
# Products with total quantity greater than 100

run_query("""
SELECT "Product Name", SUM(Quantity) AS Total_Quantity
FROM superstore
GROUP BY "Product Name"
HAVING SUM(Quantity) > 100
ORDER BY Total_Quantity DESC;
""")
```

```python
# Customers with highest sales

run_query("""
SELECT "Customer Name",sum(sales) AS Total_Sales
FROM superstore
GROUP BY "Customer Name"
ORDER BY Total_Sales DESC
LIMIT 10;
""")
```

```python
# Categorize discounts

run_query("""
SELECT
    CASE
        WHEN Discount = 0 THEN 'No Discount'
        WHEN Discount <= 0.20 THEN 'Low Discount'
        ELSE 'High Discount'
    END AS Discount_Category,
    COUNT(*) AS Orders
FROM superstore
GROUP BY Discount_Category;
""")
```

```python
# Profit by category

run_query("""
SELECT
    "Category",
    ROUND(SUM(Profit),2) AS Total_Profit
FROM superstore
GROUP BY "Category"
ORDER BY Total_Profit DESC;
""")
```

```python
conn.execute("""
CREATE TABLE customers AS
SELECT DISTINCT
    "Customer ID",
    "Customer Name",
    Segment,
    Region,
    State,
    City
FROM superstore;
""")
```

```python
conn.execute("""
CREATE TABLE products AS
SELECT DISTINCT
    "Product ID",
    "Product Name",
    Category,
    "Sub-Category"
FROM superstore;
""")
```

```python
conn.execute("""
CREATE TABLE orders AS
SELECT
    "Order ID",
    "Order Date",
    "Ship Date",
    "Ship Mode",
    "Customer ID",
    "Product ID",
    Sales,
    Quantity,
    Discount,
    Profit
FROM superstore;
""")
```

```python
# INNER JOIN
# Shows customer name along with purchased product and sales.

run_query("""
SELECT
    o."Order ID",
    c."Customer Name",
    p."Product Name",
    o.Sales
FROM orders o
INNER JOIN customers c
ON o."Customer ID" = c."Customer ID"
INNER JOIN products p
ON o."Product ID" = p."Product ID"
LIMIT 10;
""")
```

```python
# LEFT JOIN
# Displays all customers even if they have no matching orders.

run_query("""
SELECT
    c."Customer Name",
    o."Order ID",
    o.Sales
FROM customers c
LEFT JOIN orders o
ON c."Customer ID" = o."Customer ID"
LIMIT 10;
""")
```

```python
# RIGHT JOIN

run_query("""
SELECT
    o."Order ID",
    c."Customer Name",
    o.Sales
FROM orders o
LEFT JOIN customers c
ON o."Customer ID" = c."Customer ID"
LIMIT 10;
""")
```
